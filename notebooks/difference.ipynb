```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
add1=pd.read_csv('../output/Agg_SIR.csv')
add2=pd.read_csv('../output/figure/Agg_SIR.csv')
```

```python
add1.columns
```

```python
a=pd.merge(add1[['Policy4','Deaths','EmpDaysLost']],add2[['Policy4','Deaths','EmpDaysLost']],on='Policy4')
b=pd.merge(add1[['Policy4','BaselineAdjDeaths','BaselineAdjEmpDays']],add2[['Policy4','BaselineAdjDeaths','BaselineAdjEmpDays']],on='Policy4')
```

```python
a.columns
```

```python
plt.scatter(a['EmpDaysLost_x']-a['EmpDaysLost_y'],a['Deaths_x']-a['Deaths_y'])
```

```python
plt.scatter(a['EmpDaysLost_x'],a['Deaths_x'])
#plt.scatter(a['EmpDaysLost_y'],a['Deaths_y'])
```

```python
b['Work Policy']=b['Policy4'].apply(lambda x:int(x[1]))
```

```python
b['empdiff']=-b['BaselineAdjEmpDays_x']+b['BaselineAdjEmpDays_y']
b['deathdiff']=-b['BaselineAdjDeaths_x']+b['BaselineAdjDeaths_y']
```

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.scatterplot(x='empdiff',y='deathdiff',hue='Work Policy',data=b,ax=ax,palette='bright')
ax.xtitle('Relative difference in employment ')
fig.show()
```
